```python
# Create the boxplot
ax = sns.boxplot(y='ova', x='bp', data=full_data)

# Calculate descriptive statistics
stats = full_data.groupby('bp')['ova'].describe()

# Display the statistics
print(stats)
plt.show()
```

```python
# Create the boxplot without outliers
sns.boxplot(y='ova', x='bp', data=full_data, showfliers=False)

# Calculate quartiles for 'ova' column
Q1 = full_data['ova'].quantile(0.25)
Q3 = full_data['ova'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the function to filter outliers and calculate min and max values
def filter_and_calculate(group):
    # Calculate quartiles for the group
    Q1_group = group['ova'].quantile(0.25)
    Q3_group = group['ova'].quantile(0.75)
    
    # Calculate the interquartile range (IQR) for the group
    IQR_group = Q3_group - Q1_group
    
    # Define the lower and upper bounds for the group
    lower_bound_group = Q1_group - 1.5 * IQR_group
    upper_bound_group = Q3_group + 1.5 * IQR_group
    
    # Filter the group to exclude outliers
    filtered_group = group[(group['ova'] >= lower_bound_group) & (group['ova'] <= upper_bound_group)]
    
    # Calculate min and max values in the filtered group
    min_value_group = filtered_group['ova'].min()
    max_value_group = filtered_group['ova'].max()
    
    # Return the min and max values for the group
    return pd.Series({'min_value': min_value_group, 'max_value': max_value_group})

# Apply the function to each group based on 'bp' column
bounds_df = full_data.groupby('bp').apply(filter_and_calculate)

print(bounds_df)
plt.show()
```
