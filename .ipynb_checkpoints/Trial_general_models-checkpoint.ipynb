# First linear regression model

```python
# include 'ova' and the values which have positive correlation >= 0.5
'''
reactions           0.869824
base_stats          0.832967
composure           0.703798
pas                 0.700421
dri                 0.643046
value               0.632681
total_stats         0.604432
power               0.544295
phy                 0.543247
shot_power          0.535906
age                 0.519422
vision              0.508012'''

# df_metrics will be the dataframe including all the variables to be included in the linear regression model
df_metrics = full_data [['ova','reactions','base_stats','composure', 'pas','dri','value','total_stats', 'power','phy','shot_power','age', 'vision']]
df_metrics
```

```python
df_metrics.isna().sum()
```

```python
df_metrics = df_metrics.dropna()
```

```python
Y = df_metrics['ova']
X = df_metrics[['reactions','base_stats','composure', 'pas','dri','value','total_stats', 'power','phy','shot_power','age', 'vision']]
X = sm.add_constant(X)

model = sm.OLS(Y,X).fit()

model.summary()
```

```python
Y = df_metrics['ova']
X = df_metrics[['reactions','base_stats','composure', 'pas','dri','value','total_stats', 'power','phy','shot_power','age', 'vision']]

lm = linear_model.LinearRegression()
lm = lm.fit(X,Y)

lm.score(X,Y)

predictions = lm.predict(X)
predictions

print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error (Y, predictions)
mae = mean_absolute_error (Y, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```

```python
# all of these variables have P value < 0, so they can be kept for now
# R2 value can be higher and RMSE should be much lower
# in order to reduce error, we have to reduce noise from the graphs
```

# Train test split

```python
from sklearn.model_selection import train_test_split

Y = df_metrics['ova']
X = df_metrics.drop(['ova'], axis=1)
X = sm.add_constant(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = sm.OLS(Y_train,X_train).fit()

model.summary()

lm = linear_model.LinearRegression()
lm = lm.fit(X_train,Y_train)

lm.score(X_train,Y_train)

predictions = lm.predict(X_test)
predictions


print("R2 value is = ",round(r2_score(Y_test, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error (Y_test, predictions)
mae = mean_absolute_error (Y_test, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```
